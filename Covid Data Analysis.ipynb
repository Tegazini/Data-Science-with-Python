## b. Data Cleaning

### i. Renaming the columns

```python
df.columns
```

```python
#Renaming the columns and assigning it to a new variable
df1 = df.rename(columns = {'DATE ':'DATE', 'country ':'COUNTRY', ' NEW Cases':'NEW_CASES', 'vaccinated ':'VACCINATED'})
```

```python
df1
```

### ii. Handling duplication

```python
df1[df1['DATE'].duplicated() == 1]
```

```python
#Dropping the duplicated rows in the dataset and assigning to a new variable
df2 = df1.drop_duplicates().copy()
```

```python
#Confirming that the duplicated rows have been dropped
df2.loc[df.duplicated()].count()
```

```python
#Resetting the index
df2.reset_index(drop = True)
```

```python
#Checking the shape of the new variable df2
df2.shape
```

Note:\
After handling the duplication and creating a new variable, the new dataset has 150 rows and 5 columns

### iii. Handling the entries "unknown" within the dataset

```python
#Checking if there are entries with "unknown" in the DATE Column
df2.query('DATE == "unknown"')
```

```python
#Checking if there are entries with "unknown" in the COUNTRY Column
df2.query('COUNTRY == "unknown"')
```

```python
#Checking if there are entries with "unknown" in the NEW_CASES Column
df2.query('NEW_CASES == "unknown"')
```

```python
#Checking if there are entries with "unknown" in the NEW_DEATHS Column
df2.query('NEW_DEATHS == "unknown"')
```

```python
#Checking if there are entries with "unknown" in the VACCINATED Column
df2.query('VACCINATED == "unknown"')
```

Note:\
The "unknown" entries are only located in the VACCINATED column in the dataset.

```python
#Dropping all the "unknown" values in the data
df2 = df2.drop(df2[df2['VACCINATED'] == "unknown"].index)
```

```python
df2.shape
```

NB:\
After removing values with "unknown" in the dataset, the data

### v. Handling the missing values in the dataset

```python
#Calling the missing values in the data
df2.isnull().any()
```

```python
#Dropping the missing values in the data and assigning to a new variable df3
df3=df2.dropna()
df3
```

```python
df3.shape
```

```python
#Resetting the index of the new dataframe
df3 = df3.reset_index(drop = True)
```

### vi. Changing the datatypes as appropriate

```python
df3.dtypes
```

```python
#Changing DATE from string to datetime object
df3['DATE'] = pd.to_datetime(df3['DATE'], format='mixed')
#Changing VACCINATED from string to int object
df3['VACCINATED'] = (df3['VACCINATED']).astype(float)
```

```python
df3.dtypes
```

```python
df3
```

# 3. VISUALIZATION

```python
#Creating a new variable with DATE dropped
df4 = df3.drop('DATE', axis = 1)
```

```python
#Creating a variable from grouping df4 by COUNTRY
data = df4.groupby('COUNTRY')
```

```python
#Visualizing the distribution of combined covid new cases, new deaths and vaccinated during the period (November 2024) with a multiple bar chat
#plt.figure(figsize=(8,6))
ax = data.agg('sum').plot.bar(figsize=(8,5))
ax.set(title = 'Trend of new cases, new deaths and vaccination among the countries in the data.')
ax.set_ylabel('Counts')

for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2, p.get_height() + 0.5, int(p.get_height()), ha='center', rotation =0)
plt.show()
```

```python
#Creating a new variable from aggregation by mean, min, max and sum and viewing the result
new = data.agg(['mean', 'min', 'max', 'sum'])
new
```

#### Adding a new columns to the dataset

Since all the dates are in November of 2024, a new column with day of the month can be included

```python
#Adding DAY column to the dataset and viewing the result
df3['DAY'] = df3['DATE'].dt.day.astype(str)
df3
```

```python
#Calculating the mean distribution of covid new cases, new deaths and vaccinated among the five countries
new2 = data.agg('mean').round(2)

new2
```

```python
#Visualizing the mean distribution of covid new cases, deaths and vaccinated among the five countries using a stacked bar chart
ax = new2.plot(kind='bar', stacked=True, figsize=(10,6))

for i, (index_value, row) in enumerate(new2.iterrows()):
    cumulative_sum = 0
    for j, value in enumerate(row):
        if value != 0:
            ax.text(i, cumulative_sum + value/2, str(value), ha='center', va='center')
            cumulative_sum += value
plt.title('Stacked bar chat of the mean distribution of new cases, new deaths and vaccinated in the five countries covered in the data.')
plt.show()
```

```python
#Calculating the standard deviation of the NEW_CASES, NEW_DEATHS and VACCINATED columns
new3 = data.agg('std')
new3
```

```python
#Visualizing the results of the correlation of the standard deviations
ax = sns.heatmap(new3.corr(), annot=True)
plt.title('Correlating the standard deviations of Covid New Cases, New Deaths and Vaccinated')
plt.show()
```
